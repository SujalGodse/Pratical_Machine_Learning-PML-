### **Standard Scaler**

- Manually we can perform using formula : x - xbar / std

```python
df = pd.DataFrame({'Values':[3,6,5,1,6,8,3,8,9,4]})

df['StdScale'] = (df.Values-np.mean(df.Values))/np.std(df.Values,ddof=1)
df
```

### **MinMax Scaler**  

Formula : (x - min(x)) / (max(x) - min(x))

- values are between 0 and 1

- insure there is no outlier and if outliers are present then use robust scaler

```python
# by using manually
df['MinMax'] = (df.Values-min(df.Values))/(max(df.Values)-min(df.Values))
df
```

```python
# by using inbuilt calculation
from sklearn.preprocessing import MinMaxScaler
scl = MinMaxScaler()
```

### **Robust Scaler**

Formula : (x-median(col))/percentile(col,75)-percentile((col,75))  
**x - median / IQR**

```python
# by using manually
df['Robust'] = (df.Values-np.median(df.Values))/(np.percentile(df.Values,75)-np.percentile(df.Values,25))
df
```

```python
# by using in built calculation
from sklearn.preprocessing import RobustScaler
```

- StandardScaler

Centers data around 0 and scales it to unit variance using mean and standard deviation.

- MinMaxScaler

Scales data to a fixed range, usually 0 to 1, using minimum and maximum values.

- RobustScaler

Scales data using the median and interquartile range, making it resistant to outliers.

---

- StandardScaler

→ Mean & Standard Deviation

- MinMaxScaler

→ Minimum & Maximum
→ Output between 0 and 1

- RobustScaler

→ Median & IQR
→ Best for Outliers
