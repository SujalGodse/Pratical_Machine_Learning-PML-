## **Multicollinearity**

It occurs when there are certain pair of input factors which have a very high value corelation coefficient because of this they will change the effect of other variables in the models.

In order to detect Multicollinearity we can calculate VIF(Variance Inflation Factor) value for each column.

Ideally VIF should be less then 10.

```python
# corelation coefficient between all the pairs
df.corr()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

x = df.drop('Marks', axis=1)

vif_df = pd.DataFrame()
vif_df['Variable'] = x.columns
vif_df['VIF'] = [
    variance_inflation_factor(x.values, i)
    for i in range(x.shape[1])
]

print("\nVIF VALUES")
print(vif_df.sort_values('VIF', ascending=False))
```

from above output we noticed that the same factor which have a heavy or high corelation coefficient between them are also having VIF value.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

x = df.drop(['Marks','CoachingHours','StudyHours'], axis=1)

vif_df = pd.DataFrame()
vif_df['Variable'] = x.columns
vif_df['VIF'] = [
    variance_inflation_factor(x.values, i)
    for i in range(x.shape[1])
]

print("\nVIF VALUES")
print(vif_df.sort_values('VIF', ascending=False))
```

## **Standardization**

Unless and until standardization is performed L1 and L2 cannot be performed

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler

x = df.drop('Marks', axis=1)
y=df.Marks
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
x_scaled.head()
```

## **Linear Model**

```python
# To get the optimum value of alpha for Lasso regression

from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(x_scaled,y)
print(lassocv.alpha_)

from sklearn.linear_model import LinearRegression
regr1 = LinearRegression()
regr1.fit(x_scaled,y)
print(regr1.coef_)

regr1.intercept_
print(regr1.feature_names_in_)

```

## **Ridge Regression**

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(x_scaled,y)
print(ridgecv.alpha_)

ridge = Ridge(alpha=10)
ridge.fit(x_scaled,y)
print(ridge.coef_)
ridge.intercept_
print(regr1.feature_names_in_)
```
